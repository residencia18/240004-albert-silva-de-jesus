## **Principais diferenças em relação a C/C++:**

### ***Divisão padrão retorna float:***
### Em Python, a divisão de dois inteiros resulta em um número de ponto flutuante, a menos que você use o operador de divisão inteira `//`. Em C/C++, a divisão de inteiros resulta em um número inteiro, e se você quiser um resultado de ponto flutuante, pelo menos um dos operandos deve ser do tipo float.

### ***Sem restrições de tamanho em inteiros:***
### Em Python, os inteiros não têm um tamanho fixo, o que significa que você pode trabalhar com inteiros de qualquer tamanho, sem preocupações com estouro. Em C/C++, o tamanho dos inteiros é limitado pela arquitetura do sistema.

### ***Operadores compostos simplificados:***
### Python oferece operadores compostos mais expressivos, como `+=`, `-=`, etc., tornando o código mais conciso em comparação com C/C++.

### ***Exponenciação mais intuitiva:***
### O operador `**` em Python é usado para exponenciação, enquanto em C/C++, você usaria a função `pow` ou implementaria um loop para calcular a potência.

```python
'''
● Demonstre a possibilidade de representar números inteiros significativamente grandes calculando o fatorial de 30 e comparando o resultado com o maior valor inteiro que pode ser representado em
C/C++;
'''

print('\t='+'=' *30)
print('\tCálculo do Fatorial em Python:\n')

resultFatorial = 1

for i in range(1, 30 + 1):
    resultFatorial *= i
    print('\t', resultFatorial)

print('\n\tResultado do fatorial de 30 = {}'.format(resultFatorial))

print('\t='+'=' *70)
```

### Em C/C++, o tamanho do inteiro é geralmente de 4 bytes para `int` e 8 bytes para `long long int`. O valor máximo representável depende da arquitetura do sistema. Para um `int` de 4 bytes, o valor máximo seria \(2^{31} - 1\) (2147483647). Se considerarmos um `long long int` de 8 bytes, o valor máximo seria \(2^{63} - 1\).

### O valor do fatorial de 30 em Python, que é um número significativamente grande, não teria problemas em ser representado em Python. Python automaticamente usa uma quantidade de memória suficiente para lidar com números grandes.
